```python
import requests
from bs4 import BeautifulSoup
import re


URL = "https://yatanavi.org/text/turezure/index.html"

# URLからHTMLを取得
html = requests.get(URL).text
soup = BeautifulSoup(html, "html.parser")

# HTMLから各章段のURLを取得
chapter_urls = []
for j in range(244):
    if j < 10:
        num = "00" + str(j)
    elif j < 100:
        num = "0" + str(j)
    else:
        num = str(j)
    chapter_urls.append(f"https://yatanavi.org/text/turezure/k_tsurezure{num}.txt")


# 各章段のURLからHTMLを取得して、divタグのlevel2クラスにあるpタグのテキストをtxtファイルとして保存
for i, url in enumerate(chapter_urls):
    html = requests.get(url).text
    soup = BeautifulSoup(html, "html.parser")
    text = ""
    for p in soup.select("#dokuwiki__content > div > div.page.group > div:nth-of-type(4)"):
        p_tag = p.find_all("p")
        for q in p_tag:
            text += q.get_text()
            
        text = re.sub(r"\d+\)", "", text)
        text = re.sub(r"\(.*?\)", "", text)
    
    filename = f"k_tsurezure{str(i).zfill(3)}.txt"
    with open(filename, "w", encoding="utf-8") as f:
        f.write(text)
```

```python
import requests
from bs4 import BeautifulSoup
import re

# HTMLから各章段のURLを取得
chapter_urls = []
for j in range(244):
    if j < 10:
        num = "00" + str(j)
    elif j < 100:
        num = "0" + str(j)
    else:
        num = str(j)
    chapter_urls.append(f"https://roudokus.com/tsurezure/{num}.html")
    
a=0


# 各章段のURLからHTMLを取得して、divタグのlevel2クラスにあるpタグのテキストをtxtファイルとして保存
for i, url in enumerate(chapter_urls):
    print(url)
    html = requests.get(url)
    soup = BeautifulSoup(html.content, "html.parser")
    text = ""
    blockquotes = soup.find_all("blockquote")
    for b in blockquotes:
        b.decompose()
    content_inner = soup.find('div', id='content-inner')
    for p_tag in content_inner.find_all("p"):
        if not "■" in str(p_tag):
            if not ".html" in str(p_tag):
                if not "本日は徒然草より" in str(p_tag):
                    if not "音声が再生されます" in str(p_tag):
                        if not "『徒然草』全243段を原文と、現代語訳で朗読し、全段解説したものです。" in str(p_tag):
                            if not "YouTube" in str(p_tag):
                                if not "img" in str(p_tag):
                                    if not "●" in str(p_tag):
                                        text += p_tag.get_text()
                                        
    text = re.sub(r"（[^\u4e00-\u9faf]+）", "", text)
                                        
    try:
        filename = f"g_tsurezure{str(a).zfill(3)}.txt"
        with open(filename, "w", encoding="utf-8") as f:
            f.write(text)
    except:
        print("error")
    a += 1
```
